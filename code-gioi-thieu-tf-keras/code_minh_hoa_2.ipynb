```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

# Load dữ liệu MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Chuẩn hóa dữ liệu về khoảng [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Xây dựng mô hình CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile mô hình
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Huấn luyện mô hình
model.fit(x_train.reshape(-1, 28, 28, 1), y_train, epochs=5)

# Đánh giá độ chính xác của mô hình
model.evaluate(x_test.reshape(-1, 28, 28, 1),  y_test, verbose=2)
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense

# Load dữ liệu CIFAR-10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Chuẩn hóa dữ liệu về khoảng [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Xây dựng mô hình CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile mô hình
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Huấn luyện mô hình
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Đánh giá độ chính xác của mô hình
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('Độ chính xác trên tập kiểm tra:', test_acc)
```

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.preprocessing.text import Tokenizer

# Set the maximum number of words to use
max_features = 10000

# Load the data
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Convert the sequences of word indices to sequences of one-hot vectors
maxlen = 500
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)

# Build the model
model = Sequential()
model.add(Embedding(max_features, 32))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])

# Train the model
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3, batch_size=64)

# Evaluate the model
scores = model.evaluate(x_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))

# Make predictions
new_reviews = ['The movie was great!', 'The movie was terrible.']
tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(new_reviews)
sequences = tokenizer.texts_to_sequences(new_reviews)
padded_sequences = sequence.pad_sequences(sequences, maxlen=maxlen)
predictions = model.predict(padded_sequences)
print(predictions)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN
from tensorflow.keras.optimizers import RMSprop
from sklearn.model_selection import train_test_split

# Chuẩn bị dữ liệu
text_data = ['Tôi thích xem phim hành động', 'Anh ấy là một diễn viên nổi tiếng', 'Đó là một bộ phim tuyệt vời', 'Tôi không thích đọc sách', 'Người đàn ông đó đã đánh rơi ví tiền của mình']
labels = np.array([1, 1, 1, 0, 0])

# Token hóa dữ liệu
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(text_data)
sequences = tokenizer.texts_to_sequences(text_data)
word_index = tokenizer.word_index
print('Tìm thấy %s từ khác nhau trong các câu.' % len(word_index))

# Xử lý độ dài chuỗi
max_length = max([len(s) for s in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Tách tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2)

# Tạo mô hình RNN
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=32, input_length=max_length))
model.add(SimpleRNN(units=32))
model.add(Dense(units=1, activation='sigmoid'))

# Compile mô hình
model.compile(optimizer=RMSprop(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

# Huấn luyện mô hình
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Đánh giá mô hình
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Độ chính xác trên tập kiểm tra: %.2f%%' % (accuracy*100))

# Dự đoán kết quả cho câu mới
new_reviews = ['Tôi không thích xem phim kinh dị', 'Anh ấy đã giành giải thưởng', 'Bộ phim này rất tệ']
sequences = tokenizer.texts_to_sequences(new_reviews)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
predictions = model.predict(padded_sequences)
print(predictions)
```
